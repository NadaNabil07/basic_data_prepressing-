```python
import pandas as pd
import numpy as np

# Load your dataset
salaries_df = pd.read_csv('/content/Salaries.csv')
salaries_df.head()
```

```python
salaries_df.columns
```

```python
num_rows, num_columns = salaries_df.shape
data_types = salaries_df.dtypes
missing_values = salaries_df.isnull().sum()
```

Task 2: Descriptive Statistics

```python
salary_stats = salaries_df['TotalPay'].describe()
salary_range = salary_stats['max'] - salary_stats['min']
salary_std = salaries_df['TotalPay'].std()
```

Task 3: Data Cleaning

```python
# dropping rows with missing values
salaries_df_cleaned = salaries_df.dropna()
```

Task 4: Basic Data Visualization

```python
import matplotlib.pyplot as plt

# histogram for salary distribution
salaries_df['TotalPay'].plot(kind='hist', bins=20, title='Salary Distribution')
plt.show()

# pie chart for department proportions
department_proportion = salaries_df['JobTitle'].value_counts().head(10)
department_proportion.plot(kind='pie', autopct='%1.1f%%', title='Proportion of Employees in Top 10 Job Titles')
plt.show()
```

```python
# Bar chart for the top 10 Job Titles
top_job_titles = salaries_df['JobTitle'].value_counts().head(10)
top_job_titles.plot(kind='bar', color='skyblue', title='Top 10 Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Number of Employees')
plt.show()

# Bar chart for the number of employees in each year
yearly_employee_count = salaries_df['Year'].value_counts().sort_index()
yearly_employee_count.plot(kind='bar', color='lightcoral', title='Number of Employees Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Employees')
plt.show()
```

Task 5: Grouped Analysis


```python
# grouped Analysis by Job Title
grouped_by_department = salaries_df.groupby('JobTitle')['TotalPay'].mean()
```

Task 6: Simple Correlation Analysis

```python
#correlation Analysis
correlation_column = 'Year'
correlation = salaries_df['TotalPay'].corr(salaries_df[correlation_column])
```

Task 7: Scatter Plot for Correlation Analysis

```python
# scatter Plot for Correlation Analysis
plt.scatter(salaries_df[correlation_column], salaries_df['TotalPay'])
plt.title(f'Scatter Plot: Salary vs {correlation_column}')
plt.xlabel(correlation_column)
plt.ylabel('TotalPay')
plt.show()
```

Task 7: Heatmap for Correlation Analysis

```python
# Heatmap for Correlation Analysis
import seaborn as sns
correlation_matrix = salaries_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
